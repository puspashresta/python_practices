```python
from bs4 import BeautifulSoup
import requests

finaldata=[]
res=requests.get(f"https://www.worldometers.info/coronavirus/",'html.parser')
data=res.text
soup=BeautifulSoup(data,'lxml')

table_tag=soup.find('table')
table_tags=table_tag.findAll('tr')
table_tags=table_tags[0:-1] # used to select rows of table
#make list to add data
data_list=[]

for x in table_tags:
    data_list.append(x.get_text())
    
#data_list[0].split('\n')

for y in range(len(data_list)):
    finaldata.append(data_list[y][0:-1].split())
    
finaldata
```

```python
import csv

len_list=len(finaldata)
try:
    with open("Project3.csv","w") as csvFile:
        for i in range(len_list):
            w=csv.writer(csvFile)
            w.writerow(finaldata[i])
        print("Succeed")
except:
    print("Error occured")
```
